```python
def get_end_utility_list(self):
    end_utility_list = []
    for get_agreement_list in self.__get_agreement_list:
        end_utility = {}
        if get_agreement_list[0]:
            action = get_agreement_list[1]
            for j, agent in enumerate(self.__agent_list):
                end_utility[agent.get_name()] = self.__utilities.get_utility_space(j).get_utility_discounted(action.get_bid(), action.get_time_offered())
            # print("last turn:", self.__rule.get_time_now())
            end_utility["agreement_bid"] = action.get_bid().get_indexes()
            # print("agreement bid:", action.get_bid().get_indexes())
            # print("parato distance:", self.display.get_parato_distance(action))
            end_utility["parato_distance"] = self.display.get_parato_distance(action)
        else:
            for j, agent in enumerate(self.__agent_list):
                end_utility[agent.get_name()] = 0
            end_utility["agreement_bid"] = 0
            end_utility["parato_distance"] = 0
        end_utility_list.append(end_utility)
    return end_utility_list
```

```python
utility_list = jupiter.get_end_utility_list()
# utility_list = jupiter.get_get_agreement_list()
linear = []
improve = []
parato = []
for i in utility_list:
    print(i)
#     linear.append(i["LinearAgent"])
    improve.append(i["ImprovementAgent"])
    parato.append(i["parato_distance"])
```

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

 
# 折れ線グラフを出力
left = np.array(range(0, len(utility_list)))

# p1 = plt.plot(left, linear, marker='o', linewidth=2)
p2 = plt.plot(left, improve, marker='o', linewidth=2)
p3 = plt.plot(left, parato, marker='o', linewidth=2)
plt.title("This is a title")
plt.xlabel("the number of repetition")
plt.ylabel("utility value or parato distance")
plt.grid(True)
# plt.legend((p1[0], p2[0], p3[0]), ("LinearAgent", "ImprovementAgent", "parato_distance"), loc=2)
plt.legend(( p2[0], p3[0]), ( "ImprovementAgent", "parato_distance"), loc=2)
# plt.legend((p3[0]), ( "parato_distance"))
```
